### Annualize Returns to Dec

```python
crsp_m=util.annRet(crsp_m)

crsp_m.head()
```

```python
# Clean up pre merge
betadf.drop(columns=['month','year'],inplace=True)
```

```python
crsp = crsp_m.merge(betadf,how='inner',on=['permno','jdate'])
```

```python
crsp = util.lag(crsp,'beta','month')
```

```python
crsp = crsp.dropna()

# Pick top xxxx based on market cap per jdate, which is annual now
crsp = util.topX(crsp,'me', 2000)
```

```python
crsp.sort_values('lagbeta')
```

```python
df = crsp.copy()
#df = returns1.copy()

qsig = 'lagbeta'
#qsig = 'lagstockSD'

nquant = 5
df = util.splitQ(df,qsig,nquant)

temp = {}

for i in range(nquant):
    dftemp = df[df[qsig+'q']==i]
    import statsmodels.formula.api as smf

    def ols_coef(x,formula):
        return smf.ols(formula,data=x).fit().params

    gamma = (dftemp.groupby(['jdate']).apply(ols_coef,'annRet ~ lagbeta'))

    def fm_summary(p):
        s = p.describe().T
        s['std_error'] = s['std']/np.sqrt(s['count'])
        s['tstat'] = s['mean']/s['std_error']
        return s[['mean','std_error','tstat']]

    temp['q='+str(i)]=fm_summary(gamma)
    
print(str(dftemp.groupby(['jdate'])['permno'].count()[0])+' stocks in each quantile')
for entry in temp:
    print('\n'+entry,'\n',temp[entry])
```
