### Better Beta Regression w/o Quantiles

```python
df_ols = crsp_sorted.copy()

import statsmodels.formula.api as smf

def ols_coef(x,formula):
    return smf.ols(formula,data=x).fit().params

gamma = (df_ols.groupby(['jdate']).apply(ols_coef,'annRet ~ lagbeta_delta'))

def fm_summary(p):
    s = p.describe().T
    s['std_error'] = s['std']/np.sqrt(s['count'])
    s['tstat'] = s['mean']/s['std_error']
    return s[['mean','std_error','tstat']]

fm_summary(gamma)
```

### Better Beta Regression w/Quantiles

```python
df = crsp_sorted.copy()
#df = returns1.copy()

qsig = 'lagbeta_delta'
#qsig = 'lagstockSD'

nquant = 5
df = util.splitQ(df,qsig,nquant)

df.head()
```

```python
temp = {}

for i in range(nquant):
    dftemp = df[df[qsig+'q']==i]
    import statsmodels.formula.api as smf

    def ols_coef(x,formula):
        return smf.ols(formula,data=x).fit().params

    gamma = (dftemp.groupby(['jdate']).apply(ols_coef,'annRet ~ lagbeta_delta'))

    def fm_summary(p):
        s = p.describe().T
        s['std_error'] = s['std']/np.sqrt(s['count'])
        s['tstat'] = s['mean']/s['std_error']
        return s[['mean','std_error','tstat']]

    temp['q='+str(i)]=fm_summary(gamma)
    
print(str(dftemp.groupby(['jdate'])['permno'].count()[0])+' stocks in each quantile')
for entry in temp:
    print('\n'+entry,'\n',temp[entry])
```
