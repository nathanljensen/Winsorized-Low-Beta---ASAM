### Import Data

#### Win Beta

```python
winbetadf = pd.read_csv('winbeta_1.csv.gz', compression='gzip', usecols=lambda x: x.lower())

winbetadf['jdate'] = pd.to_datetime(winbetadf['jdate'])

# Rename beta to win_beta
winbetadf = winbetadf.rename(columns={'beta':'win_beta'})

winbetadf.head()
```

```python
winbetadf.dtypes
```

```python
winbetadf.shape
```

#### Normal Beta

```python
betadf = pd.read_csv('beta_1.csv.gz', compression='gzip', usecols=lambda x: x.lower())

betadf['jdate'] = pd.to_datetime(betadf['jdate'])

# Rename beta to normal_beta
betadf = betadf.rename(columns={'beta':'normal_beta'})


betadf.head()
```

```python
betadf.dtypes
```

```python
betadf.shape
```

### Merge

```python
betterbetadf = winbetadf.merge(betadf[['permno', 'jdate', 'normal_beta']], on=['permno', 'jdate'], how='left')

betterbetadf.head()
```

```python
betterbetadf.shape
```

### Plot of Betas

```python
# Scatterplot
# Data
x=betterbetadf['normal_beta']
y=betterbetadf['win_beta']

# Plot
#plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.scatter(x, y)
i=np.linspace(-10,10,11) 
plt.plot(i,i,'k-') # identity line

# Plot
plt.title('WinBeta vs Beta Scatterplot')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel("Beta OLS")
plt.ylabel("Beta Winsorized")
plt.show()
```

The theory is to only chose the stocks that are above the 45 degree line. In other words, overindexing on those betas that are disproprtionally impacted by win beta.

### Create flag for Neglected Beta

```python
betterbetadf['better_beta_flag'] = np.where(((betterbetadf.normal_beta - betterbetadf.win_beta)<1) & (betterbetadf.win_beta<betterbetadf.normal_beta), 1, 0)

betterbetadf.head()
```

```python
betterbetadf.better_beta_flag.value_counts()
```
